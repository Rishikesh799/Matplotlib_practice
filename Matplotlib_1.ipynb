```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
score=[10,20,15,20,30]
plt.plot(score)
plt.show()
```

```python
score2=[5,16,30,12]
```

```python
plt.plot(score2)
plt.show()
```

# combie two score

```python
plt.plot(np.cumsum(score))
plt.plot(np.cumsum(score2))
plt.xlim((0,10))
plt.ylim((1,200))
plt.show()
```

# for changing graph view

```python
plt.plot(np.cumsum(score),label="MI",color="g",linestyle="-",linewidth=3)
plt.plot(np.cumsum(score2),label="RR",color="m",linewidth=3)
plt.xlim((1,5))
plt.ylim((1,200))
plt.xticks(ticks=range(5),labels=np.arange(5))
plt.xticks(rotation=20)
plt.legend(loc=2)
plt.xlabel("overs")
plt.ylabel("Runs")
plt.title("Team Runs")
plt.savefig("my.jpg",dpi=350)
plt.grid()
plt.show()
```

```python
plt.plot(np.cumsum(score),label="MI",color="g",marker="o",linestyle=" ")
plt.plot(np.cumsum(score2),label="RR",color="m",marker="o",linestyle=" ")
plt.xlim((1,5))
plt.ylim((1,200))
plt.xticks(ticks=range(5),labels=np.arange(5))
plt.xticks(rotation=20)
plt.legend(loc=2)
plt.xlabel("overs")
plt.ylabel("Runs")
plt.title("Team Runs")
plt.savefig("my.jpg",dpi=350)
plt.grid()
plt.show()
```
