# comparison of two year cases

```python
covid=[100,500,300,262]
covid2=[300,1500,600,462]
state=["Goa","MH","UP","GUJ"]
```

```python
plt.bar(state,covid,color="g",width=0.2,align="edge",label="2020")
plt.bar(state,covid2,color="m",width=-0.2,label="2021",align="edge")
plt.xlabel("STATE",fontsize=15)
plt.ylabel("CASES",fontsize=15)
plt.legend()
plt.show()
```

# horizantal bar graph

```python
plt.barh(state,covid,color="g",height=0.4,align="edge",label="2020")
plt.barh(state,covid2,color="m",height=-0.4,label="2021",align="edge")
plt.xlabel("STATE",fontsize=15)
plt.ylabel("CASES",fontsize=15)
plt.legend()
plt.xticks(rotation=45,fontsize=10)
plt.yticks(rotation=45,fontsize=10)
plt.show()
```
